```python
import os
import numpy as np
import random
import sys
np.set_printoptions(threshold=sys.maxsize)
```

```python
random.seed(10)
def holes(data):                         # generating holes in a 20x20 matrix
    num = random.randint(1,10)           # no. of holes (random)
    for i in range(num):
        
        x = random.randint(0,19)                              
        y = random.randint(0,19)

        # size of hole (random)
        size_x = random.randint(1,3)                             
        size_y = random.randint(1,3)

        hole = np.full((size_x,size_y),1)             # taking 1Pa as Young's modulus for hole
       
       # position of hole (random)
        if ((x + size_x) <= 20) and ((y + size_y) <= 20) :       # "data[19:20,19;20]" switches elemnt 19,19 according to syntax (counter-intuitive)
            data[x:x+size_x,y:y+size_y] = hole
        else :
            r = random.randint(5,21)

            if ((x + size_x) > 20) :
                a = (x + size_x) - r
                b = a + size_x
            else :
                a = x
                b = x + size_x

            if ((y + size_y) > 20) :
                c = (y + size_y) - r
                d = c + size_y
            else :
                c = y
                d = y + size_y

            data[a:b,c:d] = hole
        return data
```

```python
with open('/home/shreyas/ME504/project/femcode(DL_executed)/femcode(DL_executed)/cont_ver2/prop.dat', 'r') as prop:
    old_data = prop.readlines()

def data_with_hole(data,mat):
    data_hole=np.array([]) #data with hole
    data_hole=np.append(data_hole,data[0])
    for i in range(len(data)-3):
        data_hole=np.append(data_hole,str(mat[i])+'\n')
    data_hole=np.append(data_hole,'0.33'+'\n') #poisson ratio of aluminium is 0.33
    data_hole=np.append(data_hole,data[-1:]) 
    return data_hole

for i in range(600): #number of property datasets is 600
    matrix = np.full((20, 20), 70000000000)
    mat=holes(matrix)
    linear=mat.reshape(400)
    newdata = data_with_hole(old_data, linear)
    
    with open(f'prop{i}.dat', 'w') as file:
        file.writelines(newdata)
```
